```python
import pandas as pd

def unknown_function(file_path):
    
    df = pd.read_csv(file_path)

    filtered_df = df[df['age'] > 30]

    grouped_df = filtered_df.groupby('department')['salary'].mean()

    sorted_df = grouped_df.sort_values(ascending=False)

    print(sorted_df.head())

# Example usage
unknown_function('employees.csv')
```

```python
from PIL import Image, ImageFilter

def unknown_function(image_path):
    
    with Image.open(image_path) as img:
        
        filtered_img = img.filter(ImageFilter.BLUR)

        filtered_img.save('filtered_image.jpg')

        img.show()
        filtered_img.show()

# Example usage
unknown_function('example.jpg')
```

```python
import requests
from bs4 import BeautifulSoup

def unknown_function(url):
    
    response = requests.get(url)

    soup = BeautifulSoup(response.content, 'html.parser')

    title = soup.title.string
    print(f'Title: {title}')

    links = soup.find_all('a')
    for link in links:
        href = link.get('href')
        if href:
            print(href)

# Example usage
unknown_function('https://www.example.com')
```

```python
import os

def unknown_function(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            new_name = filename.replace("old", "new")
            os.rename(os.path.join(directory, filename), os.path.join(directory, new_name))
            print(f"Renamed {filename} to {new_name}")

# Example usage
directory = 'path/to/directory'
unknown_function(directory)
```

```python
def unknown_function1(filename, data):
    with open(filename, 'wb') as file:
        file.write(data)
    print(f"Data written to {filename}")

def unknown_function2(filename):
    with open(filename, 'rb') as file:
        data = file.read()
    print(f"Data read from {filename}: {data}")
    return data

def main():
    filename = 'example.bin'
    data = b'This is some binary data'
    unknown_function1(filename, data)
    read_data = unknown_function2(filename)
    print(f"Read data: {read_data}")

# Example usage
main()
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def unknown_function(limit):
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def main():
    limit = 50
    primes = unknown_function(limit)
    print(f"Prime numbers up to {limit}: {primes}")

# Example usage
main()
```

```python
def unknown_function(list1, list2):
    merged_list = []
    i, j = 0, 0
    
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    
    return merged_list

def main():
    list1 = [1, 3, 5, 7]
    list2 = [2, 4, 6, 8]
    result = unknown_function(list1, list2)
    print(f"Merged list: {result}")

# Example usage
main()
```
